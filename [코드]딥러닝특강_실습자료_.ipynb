### 좀 더 강력한 가시화 코드
    - display_samples은 여러 이미지와 레이블을 출력해줍니다(구체적인 코드 구현 부분은 다 아실 필요 없습니다)

```python
import numpy as np
import matplotlib.pyplot as plt

def display_samples(images, labels, predictions=None):
    num_samples = len(images)
    rows = num_samples // 8  # 8개 단위로 줄을 나눕니다.
    remainder = num_samples % 8  # 나머지 개수를 계산합니다.

    if remainder > 0:
        rows += 1  # 나머지가 있으면 추가 줄이 필요합니다.

    fig, axes = plt.subplots(rows, 8, figsize=(16, 2 * rows))  # 8개 단위로 그림을 그립니다.

    errors = 0
    is_eval = False
    for i in range(rows):
        for j in range(8):  # 8개 단위로 반복합니다.
            index = i * 8 + j
            if index >= num_samples:
                break

            ax = axes[i, j] if rows > 1 else axes[j]

            ax.imshow(images[index].reshape(28, 28), cmap='gray')
            ax.axis('off')
            if len(labels.shape) == 1:
                true_label = labels[index]
            elif len(labels.shape) == 2:
                true_label = np.argmax(labels[index])
                is_eval = True
            else:
                raise ValueError(f"labels.shape={labels.shape}")
            pred_label = np.argmax(predictions[index]) if predictions is not None else true_label

            color = 'green' if true_label == pred_label else 'red'
            if true_label != pred_label:
                errors += 1

            if predictions is None:
                ax.set_title(f"Label: {true_label}", color=color)
            else:
                ax.set_title(f"Label: {true_label}, Pred: {pred_label}", color=color)

    plt.tight_layout()
    plt.show()

    if is_eval:
        print(f"정확도: {100 * (num_samples - errors) / num_samples:.2f}%")
        print(f"오검출 수량: {errors} / {num_samples}")
```

```python
display_samples(train_images[8:32],
                train_labels[8:32],
                predictions=None)
```

### 모델 구축 및 학습

```python
X = train_images
y = train_labels

x_np_array = np.array(X) #type을 맞춰줘야 함
y_onehot = tf.keras.utils.to_categorical(y) #dtype=float32

model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(units=2, input_shape=(784,), activation='sigmoid'),
    tf.keras.layers.Dense(units=10, activation='softmax')
]) #Dense(유닛수=출력수, 입력_데이터_특징_차원, ac...)

model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])


# 모델 훈련(train)
history = model.fit(x_np_array, y_onehot, epochs=3)
```
